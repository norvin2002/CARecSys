## Code 1 - 1 master csv for all hotels

```python
def bookingscrap(URL, pagecount, filename_seq):
    
    # loop to each and every review page of a hotel
    for page in range(pagecount):

        options = webdriver.ChromeOptions()
        options.add_argument("headless")

        driver = webdriver.Chrome(executable_path = 'C:/Users/nchandra/Documents/chromedriver', options = options)
        driver.get(URL)

        time.sleep(5)

        html = driver.page_source
        soup = bs(html, 'lxml')

        time.sleep(1)

        driver.close()

        # get review records
        records = []

        review_list = soup.find('ul', {'class':'review_list'})
        
        # get hotel_name
        
        if soup.find('h1', class_ = "item hotel_name").find('a', class_ = 'standalone_header_hotel_link') is not None:
            hotel = soup.find('h1', class_ = "item hotel_name").find('a', class_ = 'standalone_header_hotel_link').text.strip()
        else:
            hotel = ''
        
        # hotel overall score
        if soup.find('span', {'class':'review-score-badge'}) is not None:
            overall_score = soup.find('span', {'class':'review-score-badge'}).text.strip()
        else:
            overall_score = ''
            
        # hotel cleanliness
        if soup.find('ul', {'id':'review_list_score_breakdown'}).find('li',{'data-question':'hotel_clean'}) is not None:
            clean_score = soup.find('ul', {'id':'review_list_score_breakdown'}).find('li',{'data-question':'hotel_clean'}).find('p',{'class':'review_score_value'}).text.strip()
        else:
            clean_score = ''
            
        # hotel comfort
        if soup.find('ul', {'id':'review_list_score_breakdown'}).find('li',{'data-question':'hotel_comfort'}) is not None:
            comfort_score = soup.find('ul', {'id':'review_list_score_breakdown'}).find('li',{'data-question':'hotel_comfort'}).find('p',{'class':'review_score_value'}).text.strip()
        else:
            comfort_score = ''
            
        # hotel location
        if soup.find('ul', {'id':'review_list_score_breakdown'}).find('li',{'data-question':'hotel_location'}) is not None:
            loc_score = soup.find('ul', {'id':'review_list_score_breakdown'}).find('li',{'data-question':'hotel_location'}).find('p',{'class':'review_score_value'}).text.strip()
        else:
            loc_score = ''
            
        # hotel facilities
        if soup.find('ul', {'id':'review_list_score_breakdown'}).find('li',{'data-question':'hotel_services'}) is not None:
            facilities_score = soup.find('ul', {'id':'review_list_score_breakdown'}).find('li',{'data-question':'hotel_services'}).find('p',{'class':'review_score_value'}).text.strip()
        else:
            facilities_score = ''
            
        # hotel staff
        if soup.find('ul', {'id':'review_list_score_breakdown'}).find('li',{'data-question':'hotel_staff'}) is not None:
            staff_score = soup.find('ul', {'id':'review_list_score_breakdown'}).find('li',{'data-question':'hotel_staff'}).find('p',{'class':'review_score_value'}).text.strip()
        else:
            staff_score = ''
            
        # hotel value for money
        if soup.find('ul', {'id':'review_list_score_breakdown'}).find('li',{'data-question':'hotel_value'}) is not None:
            
            value_score = soup.find('ul', {'id':'review_list_score_breakdown'}).find('li',{'data-question':'hotel_value'}).find('p',{'class':'review_score_value'}).text.strip()
        
        else:
            value_score = ''
            
        # hotel wifi
        if soup.find('ul', {'id':'review_list_score_breakdown'}).find('li',{'data-question':'hotel_wifi'}) is not None:
            
            wifi_score = soup.find('ul', {'id':'review_list_score_breakdown'}).find('li',{'data-question':'hotel_wifi'}).find('p',{'class':'review_score_value'}).text.strip()
        else:
            wifi_score = ''

        # Loop the list of review including archived reviews
        #for i in review_list.find_all('li', {'class':['review_item clearfix ','review_item clearfix archive_item ']}):
        for i in review_list.find_all('li', class_ = 'review_item clearfix'):  

            # Clear all attribute containers at each run
            hotel_name = ''
            
            review_date = ''

            reviewer_country = ''

            reviewer_rating = ''
            
            review_count = ''

            review_header = ''

            reviewer_tag_list = ''

            review_neg_content = ''

            review_pos_content = ''

            stay = ''

            hotel_name = hotel
            
            # Review Date
            if i.find('p', {'class':'review_item_date'}) is not None:
                review_date = i.find('p', {'class':'review_item_date'}).text.strip()
            else:
                review_date = ''
                
            # Reviewer's country of origin
            if i.find('div',{'class':'review_item_reviewer'}).find('span',{'class' : 'reviewer_country'}).find('span',itemprop = 'name') is not None:
                reviewer_country = i.find('div',{'class':'review_item_reviewer'}).find('span',{'class' : 'reviewer_country'}).find('span',itemprop = 'name').text.strip()
            else:
                reviewer_country = ''
            
            # Reviewer's overall rating
            if i.find('div',{'class':'review_item_review'}).find('span', {'class' : 'review-score-badge'}) is not None:
                reviewer_rating = i.find('div',{'class':'review_item_review'}).find('span', {'class' : 'review-score-badge'}).text.strip()
            else:
                reviewer_rating = ''
                
            # Reviewer's review count
            if i.find('div',{'class':'review_item_reviewer'}).find('div', class_ = "review_item_user_review_count") != None:
                review_count = i.find('div',{'class':'review_item_reviewer'}).find('div', class_ = "review_item_user_review_count").text.strip().split()[0]
            else:
                review_count = ''
           
            # Review Header
            if i.find('div',{'class':'review_item_review'}).find('div',{'class':'review_item_header_content'}) is not None:
                review_header = i.find('div',{'class':'review_item_review'}).find('div',{'class':'review_item_header_content'}).text.strip()
            else:
                review_header = ''

            # Review tag
            review_tag_list = []

            review_tag = i.find('div',{'class':'review_item_review'}).find('ul', {'class':'review_item_info_tags'})

            for tag in review_tag.find_all('li', recursive = False):
                review_tag_list.append(tag.text.strip())

            # Negative Review
            review_neg = i.find('div',{'class':'review_item_review'}).find('div', {'class':'review_item_review_content'}).find('p', {'class':'review_neg'})

            if review_neg is not None:

                review_neg_content = review_neg.find('span', {'itemprop' : 'reviewBody'}).text.strip()

            else:
                review_neg_content = ''

            # Positive Review
            review_pos = i.find('div',{'class':'review_item_review'}).find('div', {'class':'review_item_review_content'}).find('p', {'class':'review_pos'})

            if review_pos is not None:
                review_pos_content = review_pos.find('span', {'itemprop' : 'reviewBody'}).text.strip()

            else:
                review_pos_content = ''

            # Stay date

            stay_date = i.find('div',{'class':'review_item_review'}).find('div', {'class':'review_item_review_content'}).find('p',{'class':'review_staydate'})

            if stay_date is not None:
                stay = stay_date.text.strip()
            else:
                stay = ''

            # Append each review to records
            records.append([hotel_name, review_date, reviewer_country, reviewer_rating, review_count, review_header, 
                            review_tag_list, review_neg_content, review_pos_content, stay,
                           overall_score, clean_score, comfort_score, loc_score,
                           facilities_score, staff_score, value_score, wifi_score])

        # write to csv at the end crawl for every page
        file = "C:/users/nchandra/Desktop/Booking" + ".csv"
        with open(file, "a", encoding='utf-8-sig') as f:
            writer = csv.writer(f, lineterminator = '\n')

            for item in records:
                writer.writerow(item)

        # get the next page URL and when there is no more next page, break the loop

        URL = ''
        nextpage = soup.find('div', {'class':'review_list_pagination'}).find('p',{'class': 'page_link review_next_page'}).find('a')
        if nextpage is not None:

            URL = 'https://www.booking.com/'+nextpage['href']

        else:
            break
```

```python
def bookingscraper(URL, filename_seq):
    URL = URL
    #URL = 'https://www.booking.com/reviews/sg/hotel/the-inncrowd.en-gb.html'
    #URL = 'https://www.booking.com/reviews/jp/hotel/hana.html?aid=357004;label=gog235jc-1DCAsodUIEaGFuYUgzWANoyQGIAQGYATG4AQbIAQzYAQPoAQH4AQKIAgGoAgM;sid=4889bc593fc056d6ce75c9d869dbafc3'

    options = webdriver.ChromeOptions()
    options.add_argument("headless")

    driver = webdriver.Chrome(executable_path = 'C:/Users/nchandra/Documents/chromedriver', options=options)
    driver.get(URL)

    time.sleep(5)

    html = driver.page_source
    soup = bs(html, 'lxml')

    time.sleep(1)

    driver.close()

    # Get number of page to be crawled
    import math
    
    if soup.find('p',{'class':'review_list_score_count'}) != None:
        review_count = soup.find('p',{'class':'review_list_score_count'}).text
    else:
        break
        
    pagecount = math.ceil(int(review_count.split()[2])/75)

    # loop each review page
    for page in range(pagecount):

        options = webdriver.ChromeOptions()
        options.add_argument("headless")

        driver = webdriver.Chrome(executable_path = 'C:/Users/nchandra/Documents/chromedriver', options = options)
        driver.get(URL)

        time.sleep(5)

        html = driver.page_source
        soup = bs(html, 'lxml')

        time.sleep(1)

        driver.close()

        # get review records
        records = []

        review_list = soup.find('ul', {'class':'review_list'})
        
        # get hotel_name
        
        if soup.find('h1', class_ = "item hotel_name").find('a', class_ = 'standalone_header_hotel_link') is not None:
            hotel = soup.find('h1', class_ = "item hotel_name").find('a', class_ = 'standalone_header_hotel_link').text.strip()
        else:
            hotel = ''
        
        # hotel overall score
        if soup.find('span', {'class':'review-score-badge'}) is not None:
            overall_score = soup.find('span', {'class':'review-score-badge'}).text.strip()
        else:
            overall_score = ''
            
        # hotel cleanliness
        if soup.find('ul', {'id':'review_list_score_breakdown'}).find('li',{'data-question':'hotel_clean'}) is not None:
            clean_score = soup.find('ul', {'id':'review_list_score_breakdown'}).find('li',{'data-question':'hotel_clean'}).find('p',{'class':'review_score_value'}).text.strip()
        else:
            clean_score = ''
            
        # hotel comfort
        
        if soup.find('ul', {'id':'review_list_score_breakdown'}).find('li',{'data-question':'hotel_comfort'}) is not None:
            comfort_score = soup.find('ul', {'id':'review_list_score_breakdown'}).find('li',{'data-question':'hotel_comfort'}).find('p',{'class':'review_score_value'}).text.strip()
        else:
            comfort_score = ''
            
        # hotel location
        
        if soup.find('ul', {'id':'review_list_score_breakdown'}).find('li',{'data-question':'hotel_location'}) is not None:

            loc_score = soup.find('ul', {'id':'review_list_score_breakdown'}).find('li',{'data-question':'hotel_location'}).find('p',{'class':'review_score_value'}).text.strip()
        
        else:
            loc_score = ''
            
        # hotel facilities
        if soup.find('ul', {'id':'review_list_score_breakdown'}).find('li',{'data-question':'hotel_services'}) is not None:
            facilities_score = soup.find('ul', {'id':'review_list_score_breakdown'}).find('li',{'data-question':'hotel_services'}).find('p',{'class':'review_score_value'}).text.strip()
        else:
            facilities_score = ''
            
        # hotel staff
        if soup.find('ul', {'id':'review_list_score_breakdown'}).find('li',{'data-question':'hotel_staff'}) is not None:
            staff_score = soup.find('ul', {'id':'review_list_score_breakdown'}).find('li',{'data-question':'hotel_staff'}).find('p',{'class':'review_score_value'}).text.strip()
        else:
            staff_score = ''
            
        # hotel value for money
        if soup.find('ul', {'id':'review_list_score_breakdown'}).find('li',{'data-question':'hotel_value'}) is not None:
            
            value_score = soup.find('ul', {'id':'review_list_score_breakdown'}).find('li',{'data-question':'hotel_value'}).find('p',{'class':'review_score_value'}).text.strip()
        
        else:
            value_score = ''
            
        # hotel wifi
        if soup.find('ul', {'id':'review_list_score_breakdown'}).find('li',{'data-question':'hotel_wifi'}) is not None:
            
            wifi_score = soup.find('ul', {'id':'review_list_score_breakdown'}).find('li',{'data-question':'hotel_wifi'}).find('p',{'class':'review_score_value'}).text.strip()
        else:
            wifi_score = ''

        # Loop the list of review including archived reviews
        #for i in review_list.find_all('li', {'class':['review_item clearfix ','review_item clearfix archive_item ']}):
        for i in review_list.find_all('li', class_ = 'review_item clearfix'):  

            # Clear all attribute containers at each run
            hotel_name = ''
            
            review_date = ''

            reviewer_country = ''

            reviewer_rating = ''
            
            review_count = ''

            review_header = ''

            reviewer_tag_list = ''

            review_neg_content = ''

            review_pos_content = ''

            stay = ''

            hotel_name = hotel
            
            #review_date = soup.find('ul', {'class':'review_list'}).find('p', {'class':'review_item_date'}).text.strip()
            if i.find('p', {'class':'review_item_date'}) is not None:
                review_date = i.find('p', {'class':'review_item_date'}).text.strip()
            else:
                review_date = ''
                
            # Reviewer's country of origin
            if i.find('div',{'class':'review_item_reviewer'}).find('span',{'class' : 'reviewer_country'}).find('span',itemprop = 'name') is not None:
                reviewer_country = i.find('div',{'class':'review_item_reviewer'}).find('span',{'class' : 'reviewer_country'}).find('span',itemprop = 'name').text.strip()
            else:
                reviewer_country = ''
            
            # Reviewer's overall rating
            if i.find('div',{'class':'review_item_review'}).find('span', {'class' : 'review-score-badge'}) is not None:
                reviewer_rating = i.find('div',{'class':'review_item_review'}).find('span', {'class' : 'review-score-badge'}).text.strip()
            else:
                reviewer_rating = ''
                
            # Reviewer's review count
            if i.find('div',{'class':'review_item_reviewer'}).find('div', class_ = "review_item_user_review_count") != None:
                review_count = i.find('div',{'class':'review_item_reviewer'}).find('div', class_ = "review_item_user_review_count").text.strip().split()[0]
            else:
                review_count = ''
           
            # Review Header
            if i.find('div',{'class':'review_item_review'}).find('div',{'class':'review_item_header_content'}) is not None:
                review_header = i.find('div',{'class':'review_item_review'}).find('div',{'class':'review_item_header_content'}).text.strip()

            else:
                review_header = ''

            # Review tag
            review_tag_list = []

            review_tag = i.find('div',{'class':'review_item_review'}).find('ul', {'class':'review_item_info_tags'})

            for tag in review_tag.find_all('li', recursive = False):
                review_tag_list.append(tag.text.strip())

            # Negative Review
            review_neg = i.find('div',{'class':'review_item_review'}).find('div', {'class':'review_item_review_content'}).find('p', {'class':'review_neg'})

            if review_neg is not None:

                review_neg_content = review_neg.find('span', {'itemprop' : 'reviewBody'}).text.strip()

            else:
                review_neg_content = ''

            # Positive Review
            review_pos = i.find('div',{'class':'review_item_review'}).find('div', {'class':'review_item_review_content'}).find('p', {'class':'review_pos'})

            if review_pos is not None:
                review_pos_content = review_pos.find('span', {'itemprop' : 'reviewBody'}).text.strip()

            else:
                review_pos_content = ''

            # Stay date

            stay_date = i.find('div',{'class':'review_item_review'}).find('div', {'class':'review_item_review_content'}).find('p',{'class':'review_staydate'})

            if stay_date is not None:
                stay = stay_date.text.strip()
            else:
                stay = ''

            # Append each review to records
            records.append([hotel_name, review_date, reviewer_country, reviewer_rating, review_count, review_header, 
                            review_tag_list, review_neg_content, review_pos_content, stay,
                           overall_score, clean_score, comfort_score, loc_score,
                           facilities_score, staff_score, value_score, wifi_score])

        # write to csv at the end crawl for every page
        file = "C:/users/nchandra/Desktop/Booking_" + filename_seq + ".csv"
        with open(file, "a", encoding='utf-8-sig') as f:
            writer = csv.writer(f, lineterminator = '\n')

            for item in records:
                writer.writerow(item)


        # get the next page URL and when there is no more next page, break the loop

        URL = ''
        nextpage = soup.find('div', {'class':'review_list_pagination'}).find('p',{'class': 'page_link review_next_page'}).find('a')
        if nextpage is not None:

            URL = 'https://www.booking.com/'+nextpage['href']

        else:
            break
```
